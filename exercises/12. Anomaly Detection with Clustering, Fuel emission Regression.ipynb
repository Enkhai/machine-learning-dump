## Task 1

We begin by importing all needed libraries.

We will use the following:
- pandas for data loading and preprocessing
- sklearn for data transforming, modeling and evaluation
- matplotlib and seaborn for plotting
- numpy for miscellaneous operations

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans, Birch
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns
```

### Data preparation

We load our training set.

```python
train_df = pd.read_csv('NSL-KDD/NSL-KDDTrain.csv')
train_df
```

And take a look at some info.

```python
train_df.info()
```

Then, we load our test set.

```python
test_df = pd.read_csv('NSL-KDD/NSL-KDDTest.csv')
test_df
```

And take a glimpse at some info as well.

Both sets are clean and do not have missing values.

```python
test_df.info()
```

Then, we want to one-hot encode our data.

For this we build separate sets for each OHE column.

```python
train_protocol_type_df = pd.get_dummies(train_df.protocol_type).add_prefix('protocol_type_')
train_service_df = pd.get_dummies(train_df.service).add_prefix('service_')
train_flag_df = pd.get_dummies(train_df.flag).add_prefix('flag_')
```

Then, we make our new training set and drop the old columns.

```python
train_df = pd.concat([train_df, train_protocol_type_df, train_service_df, train_flag_df], axis=1)\
                .drop(columns=['protocol_type', 'service', 'flag'])
train_df
```

We do the same thing for the test set, using the columns from the training set in case we have missing values for any OHE column type.

```python
new_cols = train_protocol_type_df.columns.append(train_service_df.columns).append(train_flag_df.columns)
```

```python
test_protocol_type_df = pd.get_dummies(test_df.protocol_type).add_prefix('protocol_type_')
test_service_df = pd.get_dummies(test_df.service).add_prefix('service_')
test_flag_df = pd.get_dummies(test_df.flag).add_prefix('flag_')
```

```python
new_test_df = pd.concat([test_protocol_type_df, test_service_df, test_flag_df], axis=1)
```

Since we may have missing columns in the test set, we make sure all columns that exist in the training set exist in the test set.

For values that do not exist for training OHE columns we make sure that for those columns the values are 0.

```python
for col in new_cols:
    if col in new_test_df.columns:
        test_df[col] = new_test_df[col]
    else:
        test_df[col] = 0

test_df
```

Then, we drop the original columns from the test set, just like we did for the training set.

```python
test_df = test_df.drop(columns=['protocol_type', 'service', 'flag'])
test_df
```

We then build our test features and targets.

The training set does not contain targets, so we will not bother with it.

```python
test_features = test_df.drop(columns='target')
test_features
```

We transform the string target values to numeric values.

```python
test_targets = test_df.target.replace({'normal':0, 'attack':1})
test_targets
```

We scale all features.

```python
scaler = StandardScaler()
train_df[:] = scaler.fit_transform(train_df)
test_features[:] = scaler.transform(test_features)
```

And perform PCA to reduce dimensions.

```python
pca = PCA(10)
train_features = pca.fit_transform(train_df)
test_features = pca.transform(test_features)
```

### Modeling and evaluation

For training we will use three different models. Gaussian mixture, KMeans and Birch.

#### Gaussian mixture

```python
mixture_clf = GaussianMixture().fit(train_features)
```

```python
mixture_predicted = mixture_clf.predict(test_features)
```

Our model did not perform well at all.

```python
print('F1:', metrics.f1_score(test_targets, mixture_predicted))
print('Accuracy:', metrics.accuracy_score(test_targets, mixture_predicted))
print('Precision:', metrics.precision_score(test_targets, mixture_predicted))
print('Recall:', metrics.recall_score(test_targets, mixture_predicted))
```

#### K-Means

```python
kmeans_clf = KMeans(n_clusters=2).fit(train_features)
```

```python
kmeans_predicted = kmeans_clf.predict(test_features)
```

KMeans did a somewhat better job.

```python
print('F1:', metrics.f1_score(test_targets, kmeans_predicted))
print('Accuracy:', metrics.accuracy_score(test_targets, kmeans_predicted))
print('Precision:', metrics.precision_score(test_targets, kmeans_predicted))
print('Recall:', metrics.recall_score(test_targets, kmeans_predicted))
```

#### Birch

```python
birch_clf = Birch(n_clusters=2).fit(train_features)
```

```python
birch_predicted = birch_clf.predict(test_features)
```

Birch did not perform well either.

```python
print('F1:', metrics.f1_score(test_targets, birch_predicted))
print('Accuracy:', metrics.accuracy_score(test_targets, birch_predicted))
print('Precision:', metrics.precision_score(test_targets, birch_predicted))
print('Recall:', metrics.recall_score(test_targets, birch_predicted))
```

## Task 2

Yet again, we begin by importing all needed libraries.

For managing our data we will use pandas, for plotting, matplotlib and seaborn, and for splitting and evaluating our data, sklearn. For our model algorithms, we will use sklearn and XGBoost.

We will also use numpy and scipy.stats for miscellaneous operations.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn import metrics
from scipy.stats import skew
```

### Data preparation

We load our dataset and start exploring.

```python
df = pd.read_csv('fuel_emissions.csv')
df
```

```python
df.info()
```

Let's start by investigating our missing values percentages.

```python
missing_df = (df.isnull().sum() / len(df)) * 100
missing_df = pd.DataFrame({'Missing Percentage': missing_df})
missing_df.sort_values(by='Missing Percentage', ascending=False, inplace=True)
missing_df['Missing Percentage'] = missing_df['Missing Percentage'].round(4)
missing_df
```

It looks like we have a lot of missing data. Let's now drop the columns with the high missing percentages.

After doing so we are also going to drop any remaining NaN values.

```python
cols_to_remove = ['file', 'model', 'description', 'transmission', # 'fuel_cost_12000_miles',
                  'standard_12_months', 'standard_6_months', 'first_year_12_months',
                 'first_year_6_months', 'thc_nox_emissions', 'tax_band', 
                  'particulates_emissions', 'thc_emissions']
df = df.drop(columns=cols_to_remove).dropna()
df
```

Some of our columns include non-numeric values.

```python
df.dtypes
```

Let's now investigate for outliers. We do that by creating a new dataframe with the skewness of each numeric column.

```python
numeric_feats = df.dtypes[df.dtypes != 'object'].index

skewed_feats = df[numeric_feats].apply(lambda x: skew(x.dropna())).sort_values(ascending=False)
skewness = pd.DataFrame({'Skew': skewed_feats})
skewness
```

NOX emissions, particulates emissions and noise level columns display a high level of skewness, indicating high probability of outliers within them.

Let's take a look at our most highly skewed columns.

```python
plt.figure(figsize=(4,4))
sns.boxplot(df['nox_emissions'])
plt.show()
plt.figure(figsize=(4,4))
sns.boxplot(df['particulates_emissions'])
plt.show()
plt.figure(figsize=(4,4))
sns.boxplot(df['noise_level'])
plt.show()
plt.figure(figsize=(4,4))
sns.boxplot(df['engine_capacity'])
plt.show()
```

We remove outliers in the noise level and NOX emissions columns.

```python
df = df.drop(df[df.noise_level < 40].index)
df = df.drop(df[df.nox_emissions > 50000].index)
```

Let's take a look at the skewness once more.

```python
numeric_feats = df.dtypes[df.dtypes != 'object'].index

skewed_feats = df[numeric_feats].apply(lambda x: skew(x.dropna())).sort_values(ascending=False)
skewness = pd.DataFrame({'Skew': skewed_feats})
skewness
```

Some columns still seem to be skewed. To combat this, we replace those columns with their log(x + 1) values. log1p, as it is known in the numpy library, adds 1 to the logarithm to avoid division by zero errors.

```python
skewed_features = skewness[skewness.Skew > 1].Skew.index.to_list()
# skewed_features.remove('co2')
skewed_features.remove('fuel_cost_12000_miles')
for feature in skewed_features:
    df[feature] = np.log1p(df[feature])
```

And we check again for skewness. Everything seems to be fine.

```python
numeric_feats = df.dtypes[df.dtypes != 'object'].index

skewed_feats = df[numeric_feats].apply(lambda x: skew(x.dropna())).sort_values(ascending=False)
skewness = pd.DataFrame({'Skew': skewed_feats})
skewness
```

We one-hot-encode our data to prepare them as input for our models.

```python
ohe = pd.get_dummies(df)
ohe.head()
```

Our data is now clean, and without having to discard many rows.

We now prepare our features and targets, which we will then feed to a linear regression, a random forest regression and an XGBoost regression model.

```python
# filtered_features = filtered_ohe.drop(columns='co2')
# filtered_targets = filtered_ohe.co2
features = ohe.drop(columns='fuel_cost_12000_miles')
targets = ohe.fuel_cost_12000_miles
```

We split into training and test sets.

```python
x_train, x_test, y_train, y_test = train_test_split(features, targets)
```

### Modeling and evaluation

#### Linear regression

We fit our first linear regression model and make some predictions.

```python
lr_regressor = LinearRegression(n_jobs=4)
lr_regressor.fit(x_train, y_train)
lr_pred = lr_regressor.predict(x_test)
```

To evaluate our model, we will check the Mean Absolute Error, Mean Absolute Percentage Error as well as the R-Squared score.

```python
print('Mean absolute error:', metrics.mean_absolute_error(y_test, lr_pred))
print('R-Squared:', metrics.r2_score(y_test, lr_pred))
```

Despite our model being a simple linear regression model, it seems to have performed decently.

The mean absolute error indicates that the model is on average ~114 points off from the true values, while the R-Squared score indicates a good line fit.

To verify this, let's run a test with one of the test points. We select point with index 874.

First, let's take a look at the point and its target value.

```python
x_test.iloc[874], y_test.iloc[874]
```

```python
lr_regressor.predict(x_test.iloc[874].to_numpy().reshape(1,-1))
```

Our model predicted a value of ~2905, which is close to the ground truth, 3150.

#### Random Forest

As a next step, we will use a random forest regression model.

```python
rf_regressor = RandomForestRegressor(n_jobs=4)
rf_regressor.fit(x_train, y_train)
rf_pred = rf_regressor.predict(x_test)
```

Let's now evaluate our model. Since this is a random forest model, we expect model performance to dramatically improve.

```python
print('Mean absolute error:', metrics.mean_absolute_error(y_test, rf_pred))
print('R-Squared:', metrics.r2_score(y_test, rf_pred))
```

The mean absolute error seems to have reduced dramatically to an average of 4.7 points off from the ground truth, while the R-Squared score has increased to an almost perfect 1.

#### XGBoost regression

As a last step, we will also perform regression using an XGBoost regression model.

```python
xgb_regressor = XGBRegressor(n_jobs=4)
xgb_regressor.fit(x_train, y_train)
xgb_pred = xgb_regressor.predict(x_test)
```

We evaluate again.

```python
print('Mean absolute error:', metrics.mean_absolute_error(y_test, xgb_pred))
print('R-Squared:', metrics.r2_score(y_test, xgb_pred))
```

Mean absolute error is somewhat higher compared to the random forest regressor with an average of ~12 point off from the true values, while the R-Squared score has is yet again an almost perfect score of 1.
