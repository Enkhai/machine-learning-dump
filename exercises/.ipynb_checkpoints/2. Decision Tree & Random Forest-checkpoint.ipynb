# Part 2: Decision Tree and Random Forest

Applying Decision Tree and Random Forest classification to scikit-learn breast_cancer dataset

We begin by importing all the libraries we are going to need.
Since we are doing classification, we select the decision tree and random forest classifiers.
We also need the datasets module to retrieve our data and metrics to measure our models' performance.

We will use the graphviz library in order to plot our tree models' architecture and save them in a human understandable format.

We will also use matplotlib for plotting the metrics of our forest classifier models.

```python
import graphviz
from sklearn import datasets, model_selection, metrics
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
```

We load our data

```python
breast_cancer_ds = datasets.load_breast_cancer()
```

And start exploring

```python
breast_cancer_ds
```

Let's take a look at our dataset's description.

```python
print(breast_cancer_ds.DESCR)
```

What is the shape of our data and the feature names?

```python
breast_cancer_ds.data.shape, breast_cancer_ds.target.shape, breast_cancer_ds.feature_names
```

We select the first 10 features and extract them, as well as the dataset's targets.

```python
num_features = 10
features = breast_cancer_ds.data[:, :num_features]
targets = breast_cancer_ds.target
features, targets
```

We then split our data in the default configuration (0.25 test set).

```python
x_train, x_test, y_train, y_test = model_selection.train_test_split(features, targets, stratify=targets)
```

## Part 1: Decision trees

For this section, we create dictionaries of 5 different models with different numbers of maximum depth, for both gini and entropy criterions.

```python
gini_models = {'gini_default_model':DecisionTreeClassifier(),
               'gini_8_model':DecisionTreeClassifier(max_depth=8),
               'gini_6_model':DecisionTreeClassifier(max_depth=6),
               'gini_5_model':DecisionTreeClassifier(max_depth=5),
               'gini_3_model':DecisionTreeClassifier(max_depth=3)
              }
```

```python
entropy_models = {'entropy_default_model':DecisionTreeClassifier(criterion='entropy'),
                  'entropy_8_model':DecisionTreeClassifier(criterion='entropy', max_depth=8),
                  'entropy_6_model':DecisionTreeClassifier(criterion='entropy', max_depth=6),
                  'entropy_5_model':DecisionTreeClassifier(criterion='entropy', max_depth=5),
                  'entropy_3_model':DecisionTreeClassifier(criterion='entropy', max_depth=3)
                 }
```

We start by training all of them using our training data.

```python
[model.fit(x_train,y_train) for model in list(gini_models.values()) + list(entropy_models.values())]
```

We then attempt to predict the targets of our test set.

```python
gini_y_predicted = {item[0] + ' test prediction':item[1].predict(x_test) for item in gini_models.items()}
entropy_y_predicted = {item[0] + ' test prediction':item[1].predict(x_test) for item in entropy_models.items()}
```

We create dictionaries for accuracy, precision, recall and F1 scores, for all of our models.

```python
gini_test_accuracy = {item[0] + ' accuracy':metrics.accuracy_score(y_test, item[1]) for item in gini_y_predicted.items()}
entropy_test_accuracy = {item[0] + ' accuracy':metrics.accuracy_score(y_test, item[1]) for item in entropy_y_predicted.items()}

gini_test_precision = {item[0] + ' precision':metrics.precision_score(y_test, item[1], average='macro') for item in gini_y_predicted.items()}
entropy_test_precision = {item[0] + ' precision':metrics.precision_score(y_test, item[1], average='macro') for item in entropy_y_predicted.items()}

gini_test_recall = {item[0] + ' recall':metrics.recall_score(y_test, item[1], average='macro') for item in gini_y_predicted.items()}
entropy_test_recall = {item[0] + ' recall':metrics.recall_score(y_test, item[1], average='macro') for item in entropy_y_predicted.items()}

gini_test_f1 = {item[0] + ' F1 score':metrics.f1_score(y_test, item[1], average='macro') for item in gini_y_predicted.items()}
entropy_test_f1 = {item[0] + ' F1 score':metrics.f1_score(y_test, item[1], average='macro') for item in entropy_y_predicted.items()}
```

And proceed to take a look at them, one by one.

```python
for gini_acc in gini_test_accuracy.items():
    print(gini_acc[0] + ': ' + str(gini_acc[1]))
print('\n')
for entropy_acc in entropy_test_accuracy.items():
    print(entropy_acc[0] + ': ' + str(entropy_acc[1]))
```

Entropy models performed better overall.

```python
for gini_prec in gini_test_precision.items():
    print(gini_prec[0] + ': ' + str(gini_prec[1]))
print('\n')
for entropy_prec in entropy_test_precision.items():
    print(entropy_prec[0] + ': ' + str(entropy_prec[1]))
```

Entropy models once again do better.

```python
for gini_rec in gini_test_recall.items():
    print(gini_rec[0] + ': ' + str(gini_rec[1]))
print('\n')
for entropy_rec in entropy_test_recall.items():
    print(entropy_rec[0] + ': ' + str(entropy_rec[1]))
```

```python
for gini_f1 in gini_test_f1.items():
    print(gini_f1[0] + ': ' + str(gini_f1[1]))
print('\n')
for entropy_f1 in entropy_test_f1.items():
    print(entropy_f1[0] + ': ' + str(entropy_f1[1]))
```

Let's take a look at how we performed on our training set.

```python
gini_y_predicted_train = {item[0] + ' train prediction':item[1].predict(x_train) for item in gini_models.items()}
entropy_y_predicted_train = {item[0] + ' train prediction':item[1].predict(x_train) for item in entropy_models.items()}
```

```python
gini_train_accuracy = {item[0] + ' accuracy':metrics.accuracy_score(y_train, item[1]) for item in gini_y_predicted_train.items()}
entropy_train_accuracy = {item[0] + ' accuracy':metrics.accuracy_score(y_train, item[1]) for item in entropy_y_predicted_train.items()}

gini_train_precision = {item[0] + ' precision':metrics.precision_score(y_train, item[1], average='macro') for item in gini_y_predicted_train.items()}
entropy_train_precision = {item[0] + ' precision':metrics.precision_score(y_train, item[1], average='macro') for item in entropy_y_predicted_train.items()}

gini_train_recall = {item[0] + ' recall':metrics.recall_score(y_train, item[1], average='macro') for item in gini_y_predicted_train.items()}
entropy_train_recall = {item[0] + ' recall':metrics.recall_score(y_train, item[1], average='macro') for item in entropy_y_predicted_train.items()}

gini_train_f1 = {item[0] + ' F1 score':metrics.f1_score(y_train, item[1], average='macro') for item in gini_y_predicted_train.items()}
entropy_train_f1 = {item[0] + ' F1 score':metrics.f1_score(y_train, item[1], average='macro') for item in entropy_y_predicted_train.items()}
```

```python
for gini_acc in gini_train_accuracy.items():
    print(gini_acc[0] + ': ' + str(gini_acc[1]))
print('\n')
for entropy_acc in entropy_train_accuracy.items():
    print(entropy_acc[0] + ': ' + str(entropy_acc[1]))
```

Overfitting becomes evident, in all cases, as we reach values of 1 with maximum depth increasing.

```python
for gini_prec in gini_train_precision.items():
    print(gini_prec[0] + ': ' + str(gini_prec[1]))
print('\n')
for entropy_prec in entropy_train_precision.items():
    print(entropy_prec[0] + ': ' + str(entropy_prec[1]))
```

```python
for gini_rec in gini_train_recall.items():
    print(gini_rec[0] + ': ' + str(gini_rec[1]))
print('\n')
for entropy_rec in entropy_train_recall.items():
    print(entropy_rec[0] + ': ' + str(entropy_rec[1]))
```

```python
for gini_f1 in gini_train_f1.items():
    print(gini_f1[0] + ': ' + str(gini_f1[1]))
print('\n')
for entropy_f1 in entropy_train_f1.items():
    print(entropy_f1[0] + ': ' + str(entropy_f1[1]))
```

We will now export our tree models' structure in pdf format, with graphviz.

```python
dot_data = {model[0] + ' tree plot': export_graphviz(model[1],
                                                     feature_names=breast_cancer_ds.feature_names[:num_features], 
                                                     class_names=breast_cancer_ds.target_names)
            for model in list(gini_models.items()) + list(entropy_models.items())}
```

We set a plot directory and render each architecture.

```python
model_plot_dir = 'model_plots/'

for data in dot_data.items():
    graph = graphviz.Source(data[1])
    graph.render(model_plot_dir + data[0])
```

## Part 2: Random forest

For forests, we want to see the metrics for forests having a different number of estimators (trees). We set an estimator step and a number of estimators and create a dictionary of multiple random forests having a different number of estimators.

```python
estimators_step = 1
n_estimators = 30
estimator_range = list(range(estimators_step, n_estimators + estimators_step, estimators_step))

gini_forests = dict([('gini_' + str(n) + '_model', RandomForestClassifier(n_estimators=n)) 
                     for n in estimator_range])
entropy_forests = dict([('entropy_' + str(n) + '_model', RandomForestClassifier(n_estimators=n, 
                                                                             criterion='entropy')) 
                     for n in estimator_range])
```

We fit each model

```python
[model.fit(x_train,y_train) for model in list(gini_forests.values()) + list(entropy_forests.values())]
```

And make predictions for our test set

```python
gini_y_predicted = {item[0] + ' test prediction':item[1].predict(x_test) for item in gini_forests.items()}
entropy_y_predicted = {item[0] + ' test prediction':item[1].predict(x_test) for item in entropy_forests.items()}
```

We calculate all metrics for our test set estimation

```python
gini_test_accuracy = {item[0] + ' accuracy':metrics.accuracy_score(y_test, item[1]) for item in gini_y_predicted.items()}
entropy_test_accuracy = {item[0] + ' accuracy':metrics.accuracy_score(y_test, item[1]) for item in entropy_y_predicted.items()}

gini_test_precision = {item[0] + ' precision':metrics.precision_score(y_test, item[1], average='macro') for item in gini_y_predicted.items()}
entropy_test_precision = {item[0] + ' precision':metrics.precision_score(y_test, item[1], average='macro') for item in entropy_y_predicted.items()}

gini_test_recall = {item[0] + ' recall':metrics.recall_score(y_test, item[1], average='macro') for item in gini_y_predicted.items()}
entropy_test_recall = {item[0] + ' recall':metrics.recall_score(y_test, item[1], average='macro') for item in entropy_y_predicted.items()}

gini_test_f1 = {item[0] + ' F1 score':metrics.f1_score(y_test, item[1], average='macro') for item in gini_y_predicted.items()}
entropy_test_f1 = {item[0] + ' F1 score':metrics.f1_score(y_test, item[1], average='macro') for item in entropy_y_predicted.items()}
```

We also make predictions for the training set

```python
gini_y_predicted_train = {item[0] + ' train prediction':item[1].predict(x_train) for item in gini_forests.items()}
entropy_y_predicted_train = {item[0] + ' train prediction':item[1].predict(x_train) for item in entropy_forests.items()}
```

And calculate all metrics for our training set estimation

```python
gini_train_accuracy = {item[0] + ' accuracy':metrics.accuracy_score(y_train, item[1]) for item in gini_y_predicted_train.items()}
entropy_train_accuracy = {item[0] + ' accuracy':metrics.accuracy_score(y_train, item[1]) for item in entropy_y_predicted_train.items()}

gini_train_precision = {item[0] + ' precision':metrics.precision_score(y_train, item[1], average='macro') for item in gini_y_predicted_train.items()}
entropy_train_precision = {item[0] + ' precision':metrics.precision_score(y_train, item[1], average='macro') for item in entropy_y_predicted_train.items()}

gini_train_recall = {item[0] + ' recall':metrics.recall_score(y_train, item[1], average='macro') for item in gini_y_predicted_train.items()}
entropy_train_recall = {item[0] + ' recall':metrics.recall_score(y_train, item[1], average='macro') for item in entropy_y_predicted_train.items()}

gini_train_f1 = {item[0] + ' F1 score':metrics.f1_score(y_train, item[1], average='macro') for item in gini_y_predicted_train.items()}
entropy_train_f1 = {item[0] + ' F1 score':metrics.f1_score(y_train, item[1], average='macro') for item in entropy_y_predicted_train.items()}
```

And then print the metrics of each model for both the test and the training sets.

```python
for gini_acc in gini_test_accuracy.items():
    print(gini_acc[0] + ': ' + str(gini_acc[1]))
print('\n')
for entropy_acc in entropy_test_accuracy.items():
    print(entropy_acc[0] + ': ' + str(entropy_acc[1]))
print('\n\n')

for gini_prec in gini_test_precision.items():
    print(gini_prec[0] + ': ' + str(gini_prec[1]))
print('\n')
for entropy_prec in entropy_test_precision.items():
    print(entropy_prec[0] + ': ' + str(entropy_prec[1]))
print('\n\n')

for gini_rec in gini_test_recall.items():
    print(gini_rec[0] + ': ' + str(gini_rec[1]))
print('\n')
for entropy_rec in entropy_test_recall.items():
    print(entropy_rec[0] + ': ' + str(entropy_rec[1]))
print('\n\n')

for gini_f1 in gini_test_f1.items():
    print(gini_f1[0] + ': ' + str(gini_f1[1]))
print('\n')
for entropy_f1 in entropy_test_f1.items():
    print(entropy_f1[0] + ': ' + str(entropy_f1[1]))
```

We define a function that will help us plot all metrics in graphs

```python
def plot_forest_metrics(criterion, metrics_list, metric_names):
    for metric, name in zip(metrics_list, metric_names):
        plt.figure(figsize=(10,6))
        plt.plot(estimator_range, metric[0].values())
        plt.plot(metric[1].values());
        plt.title(criterion + ' forests ' + name)
        plt.legend(['Test ' + name, 'Train ' + name])
        plt.xlabel('Number of estimators')
        plt.ylabel(name)
        plt.show()
```

Define lists to pass to our function

```python
gini_metrics_list = [[gini_test_accuracy, gini_train_accuracy],
               [gini_test_precision, gini_train_precision],
               [gini_test_recall, gini_train_recall],
               [gini_test_f1, gini_train_f1]]
entropy_metrics_list = [[entropy_test_accuracy, entropy_train_accuracy],
               [entropy_test_precision, entropy_train_precision],
               [entropy_test_recall, entropy_train_recall],
               [entropy_test_f1, entropy_train_f1]]
metric_names = ['Accuracy', 'Precision', 'Recall', 'F1 score']
```

And proceed to plot everything

```python
plot_forest_metrics('Gini', gini_metrics_list, metric_names)
```

```python
plot_forest_metrics('Entropy', entropy_metrics_list, metric_names)
```
