# Part 5: Naive Bayes

Applying Naive Bayes text classification on the scikit-learn newsgroups dataset

As always, we start by importing all the libraries we are going to need.

We will use the datasets module to import the 20newsgroups dataset containing documents across twenty different thematic categories, the multinomial Naive Bayes classifier for classifying these documents, and the tf-idf vectorizer to create a vocabulary, tokenize and transform tokens appearances into tf-idf scores.

We will use the metrics module to evaluate our classifier, matplotlib and seaborn for plotting and numpy for miscellaneous operations.

```python
import numpy as np
from sklearn.naive_bayes import MultinomialNB
from sklearn import datasets, metrics
from sklearn.feature_extraction.text import TfidfVectorizer
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data preparation

We load our training and testing data.

For training, we will drop all document headers, footers and quotes. This will help our classifier to generalize better.

```python
newsgroups_train = datasets.fetch_20newsgroups(subset='train', remove=('headers', 'footers,', 'quotes'))
newsgroups_test = datasets.fetch_20newsgroups(subset='test')
```

We begin to explore our data.

How is the dataset split between train and test?

### Vectorizing documents into tf-idf scores

It is now time to transform our dataset document into tf-idf token scores.

```python
vectorizer = TfidfVectorizer()
```

```python
vectors_train = vectorizer.fit_transform(newsgroups_train.data)
vectors_test = vectorizer.transform(newsgroups_test.data)
```

We also extract the targets of the training and the test set.

```python
targets_train = newsgroups_train.target
targets_test = newsgroups_test.target
```

What is the size of our vocabulary? Does it match each document?

```python
len(vectorizer.vocabulary_), vectors_train[0].shape
```

```python
print(newsgroups_train.data[0], newsgroups_train.target_names[targets_train[0]])
```

## Modeling

Let's make some models for classification. We will use 20 different models with different smoothing alpha values.

```python
alpha_step = 0.05
alpha_range = np.arange(alpha_step, 1 + alpha_step, alpha_step)
alpha_range
```

We make our models

```python
models = dict(('alpha' + str(round(alpha, 2)) + '_model', MultinomialNB(alpha=alpha)) 
              for alpha in alpha_range)
```

And train each one of them.

```python
[model.fit(vectors_train, targets_train) for model in models.values()]
```

We will use the following method for making predictions and metric values from each model, for both test and training data.

```python
def make_predictions_and_metrics():
    """
    Returns dictionaries of test and train predictions and metrics (accuracy, precision, recall and f1 
    in that order)
    """
    test_evaluations = {}
    train_evaluations = {}
    for key in models.keys():
        test_predicted = models[key].predict(vectors_test)
        test_evaluations[key] = {'predicted': test_predicted,
                                 'accuracy': metrics.accuracy_score(targets_test, test_predicted),
                                 'precision': metrics.precision_score(targets_test, test_predicted, average='macro'),
                                 'recall': metrics.recall_score(targets_test, test_predicted, average='macro'),
                                 'f1': metrics.f1_score(targets_test, test_predicted, average='macro')}
        train_predicted = models[key].predict(vectors_train)
        train_evaluations[key] = {'predicted': train_predicted,
                                  'accuracy': metrics.accuracy_score(targets_train, train_predicted),
                                  'precision': metrics.precision_score(targets_train, train_predicted, average='macro'),
                                  'recall': metrics.recall_score(targets_train, train_predicted, average='macro'),
                                  'f1': metrics.f1_score(targets_train, train_predicted, average='macro')}

    return {'test': test_evaluations, 'train': train_evaluations}
```

We calculate the predictions and metrics.

```python
evaluations = make_predictions_and_metrics()
```

We will use the following function for plotting the metrics of each model against the alpha values.

```python
def plot_metrics(evaluations):
    for metric_name in ['accuracy', 'precision', 'recall', 'f1']:
        test_values = [evaluations['test'][key][metric_name] for key in models.keys()]
        train_values = [evaluations['train'][key][metric_name] for key in models.keys()]
        
        plt.figure(figsize=(10,6))
        plt.plot(alpha_range, test_values)
        plt.plot(alpha_range, train_values)
        plt.title(metric_name.capitalize() + ' vs alpha')
        plt.legend(['Test ' + metric_name, 'Train ' + metric_name])
        plt.xlabel('alpha')
        plt.ylabel(metric_name)
        plt.show()
```

We plot.

```python
plot_metrics(evaluations)
```

We want to see how the model having an F1 score closest to 0.70 performed in the classification task. 

In this task, an F1 score much higher than 70% indicates that the model has overfitted and cannot generalize well. Such models overfit to features likes emails, keywords like `article` and specific users.

Therefore, an F1 score closer to around ~70% is much more realistic.

In this case, the model with an F1 score closer to 0.70 is the one that had an alpha value of 1.

So, to see how the model performed, we will plot the 20-class confusion matrix.

There are two ways we can do this. We can either:
1. use the already predicted values with the `confusion_matrix` function.

For this, we make values display in integer format, use the reverse rocket color map, we hide the colourbar and pass in the class labels to be displayed on the x and y ticks.

```python
plt.figure(figsize=(12,7))
sns.heatmap(metrics.confusion_matrix(targets_test, evaluations['test']['alpha1.0_model']['predicted']),
            cmap='rocket_r', annot=True, fmt='d', cbar=False,
            xticklabels=newsgroups_train.target_names, yticklabels=newsgroups_train.target_names);
```

2. use the `plot_confusion_matrix` function by passing in the classifier, the feature vectors and the targets, and calculating predictions again.

We perform similar operations in this case, this time plotting a figure on an axis and passing the axis to the function for the plot to be displayed in.

```python
plt.figure(figsize=(15,10))
ax = plt.axes()
metrics.plot_confusion_matrix(models['alpha1.0_model'], vectors_test, targets_test, 
                              display_labels=newsgroups_train.target_names,
                              xticks_rotation='vertical', values_format='d',
                              cmap='rocket_r', ax=ax);
```

Looking at the heatmap, our model seems to have performed quite well.
