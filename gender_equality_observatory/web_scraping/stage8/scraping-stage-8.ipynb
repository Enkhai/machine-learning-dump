Handling failed coauthors

```python
import json
from scholarly import scholarly
```

```python
with open('../stage7/coauthors.json') as f:
    coauthors = json.load(f)
    
with open('../stage6/uni_authors.json') as f:
    authors = json.load(f)
```

```python
coauthors['failed']
```

`DEgZaUoAAAAJ` is a professor named Nitish Thakor that has a new id `SB_7Bi0AAAAJ`. <br>
`PYu6uf0AAAAJ` is professor Ahmed Al-Tahmeesschi that no longer has a Google Scholar profile.

```python
thakor_old = 'DEgZaUoAAAAJ'
thakor_new = 'SB_7Bi0AAAAJ'

tahmeesschi = 'PYu6uf0AAAAJ'
```

What are the keys of each json?

```python
authors[0].keys(), coauthors.keys()
```

Remove Tahmeesschi from the coauthors of each author and coauthor and remove the old id of Thakor and insert the new id of Thakor, if necessary.

```python
count_for_tahmeesschi = 0
count_for_thakor = 0

for uni in authors:
    for auth in uni['authors']:
        if tahmeesschi in auth['coauthors']:
            auth['coauthors'].remove(tahmeesschi)
            count_for_tahmeesschi += 1
        if thakor_old in auth['coauthors']:
            auth['coauthors'].remove(thakor_old)
            auth['coauthors'].append(thakor_new)
            count_for_thakor += 1

for auth in coauthors['successful']:
    if tahmeesschi in auth['coauthors']:
        auth['coauthors'].remove(tahmeesschi)
        count_for_tahmeesschi += 1
    if thakor_old in auth['coauthors']:
        auth['coauthors'].remove(thakor_old)
        auth['coauthors'].append(thakor_new)
        count_for_thakor += 1
```

```python
count_for_tahmeesschi, count_for_thakor
```

Requery Thakor and append to coauthors

```python
author = scholarly.search_author_id(thakor_new)
author = scholarly.fill(author, sections=['indices', 'coauthors', 'publications'])
author
```

```python
author_dict = {'scholar_id': author['scholar_id'],
                           'name': author['name'],
                           'affiliation': author['affiliation'],
                           'url_picture': author['url_picture'],
                           'i10index': author['i10index'],
                           'i10index5y': author['i10index5y'],
                           'hindex': author['hindex'],
                           'hindex5y': author['hindex5y'],
                           'citedby': author['citedby'],
                           'citedby5y': author['citedby5y'],
                           'num_publications': len(author['publications']),
                           'coauthors': [coauth['scholar_id'] for coauth in author['coauthors']]
                           }
author_dict
```

```python
coauthors['successful'].append(author_dict)
```

Remove Thakor and Tahmeesschi from the failed list.

The list should now be empty.

```python
coauthors['failed'].remove(tahmeesschi)
coauthors['failed'].remove(thakor_old)
coauthors['failed']
```

Save authors and coauthors

```python
with open('uni_authors.json', 'w') as f:
    json.dump(authors, f)

with open('coauthors.json', 'w') as f:
    json.dump(coauthors, f)
```
