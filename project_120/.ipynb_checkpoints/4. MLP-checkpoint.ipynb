# Multi-Layer Perceptron

Problem: Classification - Digit recognition of digits 5,6,8,9 in bitmap / binary(-1,1) form

For this problem we use the Multi-Layer Perceptron (MLP) algorithm.

MLP characteristics:
- Can solve / approximate non-linearly separable data
- Comprises multiple layers of McCulloch-Pitts neurons
- Supervised
- Typically, the layers are split into:
    - One input layer
    - One ore more hidden layers
    - One output layer
- The activation of each hidden neuron is passed through a continuous, differentiable and non-linear output function. Functions:
    - Sigmoid / Logistic
    ![Sigmoid](https://www.googleapis.com/download/storage/v1/b/kaggle-forum-message-attachments/o/inbox%2F5152157%2F366ce717ad74dfe967a940d9a2161f9b%2Fmain-qimg-6b67bea3311c3429bfb34b6b1737fe0c.jpg?generation=1605711218576073&alt=media)
    - Hyperbolic tangent (tanh)
    ![Tanh](https://www.medcalc.org/manual/functions/tanh.png)
    - REctified Linear Unit (RELU)
    ![RELU](https://static.packt-cdn.com/products/9781838989590/graphics/assets/0ba4fa4a-02c5-44d3-8a81-a57fce3a21f5.png)
- The MLP can work as a Universal Approximator
- Architecture example of MLP
![MLP](https://www.researchgate.net/profile/Allan-Kardec-Barros-Filho/publication/274240858/figure/fig1/AS:392021136166914@1470476530215/TOPOLOGY-OF-A-MULTILAYER-PERCEPTRON-NEURAL-NETWORK.png)
- Inference is done through the output layer. At the output layer a softmax function is applied, with the goal of handling multiple targets
\begin{equation}
s(x_i) = \frac{e^{x_i}}{\sum_{j=1}^{n}e^{x_j}}
\end{equation}

Training:
- Initialize weights W to random small values
- For each example, the activation and the output are calculated for each neuron of the hidden layers and the output layer
- Weights are adjusted with the use of a learning rate until the MSE is minimized or a maximum number of epochs is reached.<br>
Weights change using the Back-Propagation algorithm and the derivative chain rule, adjusted towards the opposite direction of their derivatives (gradient descent).
- Due to the use of derivatives, output functions of neurons must be continuous and differentiable, therefore we avoid the use of discontinuous functions, such as the step or sign function, and instead use functions such as the sigmoid, tanh and RELU functions we saw above.

------------------

Model architecture:
- Input layer:
    - 77 neurons - as many as the digit image pixels
- Hidden layers:
    - 2 layers
    - 100 neurons at each layer
- Output layer:
    - 4 neurons - as many as classes
- Solver: Adam
- Output function of hidden layers: tanh
- Learning rate: 0.0001 - default

For this problem we are first going to the digits.<br>
This is a separate section of the problem, handled in the [Create_digits notebook](http://localhost:8888/notebooks/Create_digits.ipynb).<br>
The data we are going to create with this notebook are going to be used again for our Hopfield model in notebook no. 6.

```python
import numpy as np
from sklearn.neural_network import MLPClassifier
import matplotlib.pyplot as plt
```

We load the perfect and imperfect digits

```python
perfect_digits = np.load('perfect_digits.npy')
imperfect_digits = np.load('imperfect_digits.npy')
```

Let's see an example of a perfect digit

```python
plt.imshow(perfect_digits[0][:77].reshape(11,7), cmap='Greys')
```

And an example of an imperfect digit

```python
plt.imshow(imperfect_digits[0][:77].reshape(11,7), cmap='Greys')
```

We use the perfect and 7 imperfect digits, for each digit, for training, and 3 imperfect digits for each digit, for testing.

```python
imperfect_train_idxs = list(range(4)) + [4 + i + 9 * j for j in range(4) for i in range(6)]
imperfect_test_idx = [4 + i + 9 * j for j in range(4) for i in range(6,9)]
```

```python
X_train = np.concatenate((perfect_digits[:,:77], imperfect_digits[imperfect_train_idxs,:77]))
X_test = imperfect_digits[imperfect_train_idxs,:77]
t_train = np.concatenate((perfect_digits[:,77], imperfect_digits[imperfect_train_idxs,77]))
t_test = imperfect_digits[imperfect_test_idx,77]
```

We create our model

```python
model = MLPClassifier(hidden_layer_sizes=(100, 100), activation='tanh')
```

We also create yet another plot_training method, similar to the Perceptron and Adaline methods, which will help us understand how well our model learns the different digits, using 4 dynamic plots.

Just as before, this method will also train our model.

```python
def plot_training(model, X_train, t_train):
    fig, axes = plt.subplots(4, 1)

    for ax in axes:
        ax.set_xlim(-0.2, 1.1)
        ax.set_ylim(-0.2, 1.1)

    unique_classes = [5, 6, 8, 9]
    n_iter = 100
    items_per_class = len(t_train) // 4
    
    for i in range(n_iter):
        for ax in axes:
            ax.clear()

        axes[0].set_title('Epoch ' + str(i + 1) + ' out of ' + str(n_iter))
        model = model.partial_fit(X_train, t_train, unique_classes)

        train_pred = model.predict(X_train)
        
        for i in range(4):
            indices = np.where(t_train == unique_classes[i])
            axes[i].scatter(indices, train_pred[indices], c='r', marker='o')
            axes[i].scatter(indices, t_train[indices], c='b', marker='.')
            
        fig.canvas.draw()
        plt.pause(0.1)
        
    return model
```

We train our model, displaying in the meantime the plots.

```python
%matplotlib notebook
model = plot_training(model, X_train, t_train)
```

Finally, we display another 4 plots which will help us understand how well our models performs on the test set, for each digit.

```python
test_pred = model.predict(X_test)

labels = [5, 6, 8, 9]
items_per_class = len(t_test) // 4
for i in range(4):
    indices = np.array([j + items_per_class * i for j in range(items_per_class)])
    plt.figure(figsize=(6,6))
    plt.scatter(indices, test_pred[indices], c='r', marker='o')
    plt.scatter(indices, t_test[indices], c='b', marker='.')
    plt.title(labels[i])

    plt.show()
```
