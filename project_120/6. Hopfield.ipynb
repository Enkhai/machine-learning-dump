# Hopfield

Problem: Classification - Digit recognition of digits 5,6,8,9 in bitmap / binary(-1,1) form

For this problem we use the Hopfield algorithm:

Hopfield characteristics:
- Recurrent network
- Serves as a heteroassociative memory
- Neurons are structured into a one-dimensional grid
- Each neuron is connected with itself and to other neurons
- The neuron activation is the sum of the inputs multiplied by the corresponding synapse weights
- The output of each neuron is defined by a step function
- Hopfield network example
![Hopfield](hopfield.jpg)
- At each iteration of the inference, the output of the neurons is used as input for the next iteration
- There are multiple states of equilibrium defined by the equilibrium vectors v (attractors)
- If v is initialized in the region A<sub>1</sub>, the network will converge to the equilibrium vector  v<sub>1</sub>, if initialized in region A<sub>2</sub>, the network will converge to the equilibrium vector v<sub>2</sub>, and so on.<br>
Regions A<sub>1</sub>, A<sub>2</sub>, ..., A<sub>n</sub> are called domains of attraction.
- Types of inference:
    - Synchronous: At each epoch all neurons produce an output simultaneously
    - Asynchronous: At each epoch neurons produce an output in random order. These outputs are consequently used as input to others neurons during the same epoch.
- Energy:
\begin{equation}
E = - \frac{1}{2} \sum_{i=1}^{n}\sum_{j=1}^{n}w_{ij}v_iv_j \;Î®\; E = - \frac{1}{2} \sum_{i=1}^{n}\sum_{j=1}^{n}w_{ij}v_iv_j - \sum_{i=1}^{n}I_iv_i
\end{equation}
With each new output energy is reduced
- Storing of states is done by defining the weights of the network as the sum of the multiplications of vectors of states by their transpose vectors.<br>

```python
import numpy as np
from neurodynex3.hopfield_network import network
import matplotlib.pyplot as plt
```

This time, instead of a plot_training method, we create a plot_inference method, which will help us visualize the inference process for imperfect digits.<br>
Using plot_inference we practically call the inference for our model for a given imperfect digit and a number of iterations.<br>
Network modeling was done with the use of the neurodynex3 library, which does not allow for interactive inference, but instead returns an array with the state of the model for each iteration (run_with_monitoring(n_iter)).<br>
Using this array, we then display a plot with the provided imperfect digit, and another dynamic plot for the inference progress.

```python
def plot_inference(model, test_sample, sync=False):
    if sync:
        model.set_dynamics_sign_sync()
    else:
        model.set_dynamics_sign_async()
    
    fig = plt.figure(figsize=(8, 4))
    fig.add_subplot(121)
    fig.add_subplot(122)

    axes = fig.get_axes()

    n_iter = 10

    # Plot 1 - static
    axes[0].imshow(test_sample.reshape(11, 7), cmap='Greys')
    axes[0].set_title('Provided sample')

    pred = model.run_with_monitoring(n_iter)

    for i in range(len(pred)):
        axes[1].clear()

        # Plot 2 - dynamic
        axes[1].set_title('Iteration no. ' + str(i) + ' out of ' + str(n_iter))
        axes[1].imshow(pred[i].reshape(11,7), cmap='Greys')

        fig.canvas.draw()
        plt.pause(0.1)
    else:
        if i + 1 < n_iter:
            print('Model converged at iteration no.', str(i))
```

We load the perfect and imperfect digits we created earlier

```python
perfect_digits = np.load('perfect_digits.npy')
imperfect_digits = np.load('imperfect_digits.npy')
```

We display 2 plots, one for a perfect digit and one for an imperfect one

```python
plt.imshow(perfect_digits[0][:77].reshape(11,7), cmap='Greys')
```

```python
plt.imshow(imperfect_digits[0][:77].reshape(11,7), cmap='Greys')
```

We create our model

```python
model = network.HopfieldNetwork(11*7)
```

We store the 4 perfect digits

```python
model.store_patterns(perfect_digits[:,:77])
```

We ask the user to select between synchronous and asynchronous inference execution

```python
while True:
    answer = input('Would you like the network to run asynchronously? y/n ')
    if answer in ['y', 'n']:
        sync = answer == 'n'
        break
    print('Wrong input')
```

And finally, we select randomly one of the imperfect digits for inference and displaying the plots.

The selection between synchronous or asynchronous execution the user has previously selected is passed into the plot_inference function, calling the corresponding method which will make our model execute synchronously or asynchronously.

```python
%matplotlib notebook
test_index = np.random.choice(imperfect_digits.shape[0], replace=False)
test_sample = imperfect_digits[test_index,:77]

plot_inference(model, test_sample, sync)
```
