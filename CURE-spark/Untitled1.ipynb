```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import table
```

```python
src = 'results/noOutliers/scalability/'
```

```python
running_time = pd.read_csv(src + 'running_time.csv')
running_time
```

```python
fig, axes = plt.subplots(2, figsize=(12,10))
axes[0].xaxis.set_visible(False)
axes[0].yaxis.set_visible(False)
axes[0].set_frame_on(False)
table(axes[0], running_time, loc='center')
running_time.plot.bar(x='sample_size', ax=axes[1], logy=True)
axes[1].set_title('Time taken for different sample size in seconds (log scale)')
axes[1].legend(title='Number of partitions')
plt.show()
```

```python
fig, axes = plt.subplots(1, figsize=(10, 7))
axes.xaxis.set_visible(False)
axes.yaxis.set_visible(False)
axes.set_frame_on(False)
table(axes, running_time, loc='center')
```

```python
fig, axes = plt.subplots(1, figsize=(16, 10))
running_time[['1','2','4','6','12']].plot.bar(ax=axes, logy=True)
plt.legend(title='Number of partitions', fontsize='xx-large')
plt.show()
```

```python
running_time.groupby('sample_size').plot.bar(x='sample_size');
```
